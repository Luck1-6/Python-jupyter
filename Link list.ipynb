```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
head=Node(10)
head.next=Node(20)
head.next.next=Node(30)
head.next.next.next=Node(40)
```

```python
def traverse(head):
    p=head
    while p!=None:
        print(p.data)
        p=p.next
traverse(head)
```

```python
def sum_data(head):
    p=head
    sum=0
    while p!=None:
        sum=sum+(p.data)
        p=p.next
    return sum
    print(sum)
sum_data(head)    
```

```python
def max_data(head):
    p=head
    max=0
    while p!=None:
        if (p.data)>max:
            max=p.data
            p=p.next
    return max
    print(max)
max_data(head)        
```
