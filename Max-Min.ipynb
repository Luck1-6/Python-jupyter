```python
number1=int(input('Enter first integer: '))
```

```python
number2=int(input('Enter second integer: '))
```

```python
number3=int(input('Enter third integer: '))
```

```python
minimum=number1
```

```python
if number2<minimum:
    minimum=number2
```

```python
if number3<minimum:
    minimum=number3
```

```python
print('Minimum value is', minimum)
```
