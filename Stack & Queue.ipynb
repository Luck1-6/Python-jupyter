```python
class Stack:
    def __init__(self):
        self.values=[]
    
    def push(self, x):
        self.values.append(x)
    
    def pop(self):
        return self.values.pop()  
    def peek(self):
        return self.values[-1]

s=Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
print(s.values) 
    
print(s.pop())    
print(s.values)

s.push(60)
print(s.pop())
print(s.peek())
print(s.peek())
```

```python
class Queue:
    def __init__(self):
        self.values=[]

    def enqueue(self,x):
        self.values.append(x)

    def dequeue(self):
        try:
            return self.values.pop(0)
        except IndexError as ex:
            raise ex
q1=Queue()
q1.enqueue(10)
q1.enqueue(20)
q1.enqueue(30)
print(q1.values)

q1.enqueue(40)
print(q1.values)

print(q1.dequeue())
print(q1.values)
print(q1.dequeue())
print(q1.dequeue())
print(q1.dequeue())
print(q1.dequeue())
```
