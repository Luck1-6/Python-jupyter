```python
def _lucky():
    a,b=map(int,input().split())
    if a==9 or b==9 or abs(a+b)==9 or abs(a-b)==9:
        return ("true")
    else:
        return ("false")
_lucky()        
```

```python
def max_score():
    a,b,c=map(int,input().split())
    if (a>b)and(a>c):
        return (a)
    elif (b>c):
        return (b)
    else:
        return (c)
max_score()        
```

```python
n = int(input())
colors = input().split()
yellow = colors.count('yellow')
green = colors.count('green')
red = colors.count('red')
# Find the max count
max_count = max(yellow, green, red)
# Check how many colors have the same max count
count_max = 0
if yellow == max_count:
    count_max += 1
if green == max_count:
    count_max += 1
if red == max_count:
    count_max += 1
# Output the result
if count_max == 1:
    if yellow == max_count:
        print('yellow')
    elif green == max_count:
        print('green')
    else:
        print('red')
else:
    print('none')
```

```python
def min_moves_to_equal_strings(S, T):
    i = len(S) - 1
    j = len(T) - 1
    # Find longest common suffix
    while i >= 0 and j >= 0 and S[i] == T[j]:
        i -= 1
        j -= 1
    # The rest of the characters are not in the common suffix and must be deleted
    return (i+1)+(j+1)
# Input
S = input().strip()
T = input().strip()
# Output
print(min_moves_to_equal_strings(S, T))
```

```python
st=input()
vowel='a','e','i','o','u'
n=len(st)

if st[-1]!='n' and st[-1] not in vowel:
    print("No")
    exit()
else:
    for i in range(n):
      if st[i]!='n' and st[i] not in vowel:
          if st[i+1] not in vowel:
              print("No")
              exit()
print("Yes")
```
