# Sum of digits of a number

```python
def getSum(n): 
   
    sum = 0
    while (n != 0): 
      
        sum = sum + (n % 10)
        n = n//10
      
    return sum 
```

```python
n=int(input('integer: '))
print(getSum(n))
```

# LCM of numbers

```python
def LCM(a, b):
    greater = max(a, b)
    smallest = min(a, b)
    for i in range(greater, a*b+1, greater):
        if i % smallest == 0:
            return i
a=3
b=5
print("LCM of", a, "and", b, "is", LCM(a, b))
```

# Reverse of a number
